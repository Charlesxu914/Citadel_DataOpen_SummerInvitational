```python
city_list = num_listings['city'].unique()

for city in city_list:
  city_listings = num_listings.loc[num_listings['city'] == city]
  plt.figure()
  plt.plot(city_listings['date'].apply(lambda x: datetime.strptime(x,'%Y-%m-%d')) \
           ,city_listings['num_listings'])
  plt.title('number of listings in {} over time'.format(city))
  plt.show()
```

```python
Unemployment = unemployment_rate.rename(columns={'date':'Year_month','City':'city'})
```

```python
city = ['Los Angeles', 'Austin', 'Nashville',  'New Orleans', 'Asheville']
hpi = housing_price_index[housing_price_index['RegionName'].isin(city)].iloc[0:5,1:]
hpi = hpi.drop(hpi.columns[[i for i in range(1,164)]], axis=1)
hpi = hpi.set_index('RegionName')
hpi = pd.DataFrame(hpi.unstack().reset_index())
hpi.columns = ['date','city','housing price index']
hpi['Year_month'] = hpi.loc[:,'date'].apply(lambda x: x[:7])
hpi = hpi.iloc[:,1:]
```

```python
df = pd.merge(gdp, personal_income,on=['date','city'])
df = pd.merge(df, population,on=['date','city'])
num_listings['DATE'] = num_listings['date'].apply(lambda x: x[:4])
df['DATE'] = df.loc[:,'date'].apply(lambda x: x[:4])
```

```python
df = pd.merge(num_listings, df, on=['city','DATE'],how='inner')
df = df.drop(columns=['DATE', 'date_y'])
df = df.rename(columns={'date_x':'date'})
```

```python
df.loc[df['city'] == 'asheville' , 'city'] = 'Asheville'
df.loc[df['city'] == 'new_orleans' , 'city'] = 'New Orleans'
df.loc[df['city'] == 'nashville' , 'city'] = 'Nashville'
df.loc[df['city'] == 'austin' , 'city'] = 'Austin'
df.loc[df['city'] == 'la' , 'city'] = 'Los Angeles'
```

```python
airport_monthly = airport_monthly.rename(columns={'City':'city'})
df['Year_month'] = df.loc[:,'date'].apply(lambda x: x[:7])
```

```python
df = pd.merge(df,airport_monthly, on=['city','Year_month'],how='inner')
df = pd.merge(df,Unemployment, on=['city','Year_month'],how='inner')
df = pd.merge(df, hpi, on=['city','Year_month'],how='inner')
df = df.drop(['Year_month'], axis=1)
```

```python
df = df.sort_values(by=['date','city']).reset_index().drop(columns=['index'])
df['num_listings'] = df['num_listings']/df['population']
df = df.drop(['population'], axis=1)
df['num_listings'].replace([np.inf,-np.inf],0,inplace=True)
```

```python
def get_log_hpi(df, city):
  df_city = df[df['city'] == city].copy()
  df_city['log HPI'] = np.log(df_city['housing price index'])
  df_city = df_city.drop(['city','housing price index'], axis=1)
  df_city['date'] = pd.to_datetime(df_city['Year_month'])
  df_city = df_city[df_city['date'] < "2020-01-01"]

  baseline = df_city[df_city['Year_month'] == '2011-06']['log HPI'].iloc[0]
  std = np.std(df_city['log HPI'])
  df_city['log HPI'] = (df_city['log HPI'] - baseline)/std
  df_city = df_city.rename(columns = {'log HPI': city})
  df_city = df_city.drop(columns = 'date')
  return df_city
```

```python
hpi_log_la = get_log_hpi(hpi,'Los Angeles')
hpi_log_NO = get_log_hpi(hpi, 'New Orleans')
hpi_log_NA = get_log_hpi(hpi, 'Nashville')
hpi_log_Aus = get_log_hpi(hpi, 'Austin')
hpi_log_Ash = get_log_hpi(hpi, 'Asheville')
```

```python
ax = hpi_log_Ash.plot(x='Year_month', y='Asheville', colormap='Dark2', figsize=(14, 7))
ax.set_xlabel('Year')
ax.set_ylabel('log House Price Index (baseline: Jan 2012)')
ax.set_title('Zillow Home Value Index')

hpi_log_NO.plot(x='Year_month', y='New Orleans', ax=ax)
hpi_log_NA.plot(x='Year_month', y='Nashville', ax=ax)
hpi_log_Aus.plot(x='Year_month', y='Austin', ax=ax)
hpi_log_la.plot(x='Year_month', y='Los Angeles', ax=ax)
plt.grid()
plt.show()
```
